```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcdefaults()
fig, ax = plt.subplots(figsize=(8,4))

# Example data
models = ['APGD','MAR-ERA20C','MAR-ERA5','MAR-ECearth3','MAR-MPI','EURO-CORDEX']
years = np.array([[1970,2008],[1902,2010],[1980,2020],[1961,2015],[2015,2100],[1961,2015],[2015,2100],[2015,2100],[1950,2005],[2005,2100]])
labels = ['Observation','Reanalysis',None,'Historical','RCP 4.5',None,None,'RCP 8.5',None,None]
colors = ['C4','C0','C0','C2','C1','C2','C1','C3','C2','C3']
y_pos = np.array([0,1,2,3,3,4.5,4,5,6,6])
starts = years[:,0]
widths = years[:,1]-years[:,0]

for i in range(len(years)):
    ax.barh(y_pos[i],width = widths[i],left=starts[i], align='center',label=labels[i],color=colors[i])
# ax.barh(y_pos,width = widths,left=starts, align='center')
ax.set_yticks([0,1,2,3,4.5,6])
ax.set_yticklabels(models)
ax.invert_yaxis() 
ax.set_xlabel('Years')
# ax.set_title('Summary of all the data used')
ax.grid(color='k', linestyle='--', linewidth=0.3,axis='x')
plt.legend()
plt.show()
```
