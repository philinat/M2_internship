```python
from windrose import WindroseAxes
def plot_monthrose(fig,rect,event_months,event_maxs,title="",max_plot_rain = 200):
    axe = WindroseAxes.from_ax(fig=fig,rect=rect)
    axe.bar(event_months, event_maxs,normed=True,nsector=12,opening=0.9, bins=np.linspace(0,max_plot_rain,21),cmap= plt.get_cmap('YlGnBu'))
    # axe.set_title(title,y=0.85)
    axe.set_yticklabels([])
    axe.set_rgrids([],angle=0)
    axe.set_thetagrids(range(0,360,30),labels=range(1,13))#('J','F','M','A','M','J','J','A','S','O','N','D'))
    axe.tick_params(pad=10)
```

```python
fig = plt.figure()
rect = [0,0,1,1]
plot_monthrose(fig,rect,wd,ws,title="lama")
```
